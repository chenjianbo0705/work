```python
from tqdm import tqdm
new_dict = {'title':[],'article':[]}
for w in tqdm(range(0,len(news_df))):
    try: 
        news_url = news_df.iloc[w,2]
        res = requests.get(news_url)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        new = soup.find('div','article')
        t = soup.find('h1','main-title')
        str1=new.text.strip('\n').strip('\u3000')
        
    except:
        print('地址为{}新闻抓取失败'.format(news_url))
        continue
    
    new_dict['title'].append(t)
    new_dict['article'].append(str1)
```

```python
def getCrData(page):
    r = requests.get('http://mil.news.sina.com.cn/roll/index.d.html?cid=57918&amp;page={page}'.format(page =page))
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text,'html.parser')
    new =soup.find_all('li')
    news_dict = {'title':[],'href':[],'time':[]}
    for i in new[13:]:
        data = i.select('span')[0].text.strip('()')
        title = i.select('a')[0].text
        href =  i.select('a')[0]['href']
        news_dict['title'].append(title)
        news_dict['href'].append(href)
        news_dict['time'].append(data)
    news_df = pd.DataFrame(news_dict)[['title','time','href']]
    new_dict = {'title':[],'article':[]}
    for w in tqdm(range(0,len(news_df))):
    try: 
        news_url = news_df.iloc[w,2]
        res = requests.get(news_url)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        new = soup.find('div','article')
        t = soup.find('h1','main-title')
        str1=new.text.strip('\n').strip('\u3000')
        
    except:
        print('地址为{}新闻抓取失败'.format(news_url))
        continue
    
    new_dict['title'].append(t)
    new_dict['article'].append(str1)
    data=pd.merge(news_df,new_df,on = 'title',how = 'right')
```
