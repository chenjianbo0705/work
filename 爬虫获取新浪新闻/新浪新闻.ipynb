```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://mil.news.sina.com.cn/roll/index.d.html?cid=57918')
res.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(res.text,'lxml') 
```

```python
for i in range(13,len(news)):
    h2 = news[i].select('a')[0].text
    href = news[i].select('a')[0]['href']
    time = news[i].select('span')[0].text.strip('()')
    print(h2,href,time)
```

```python
news_dict = {'title':[],'href':[],'time':[]}

for i in range(13,len(news)):
    h2 = news[i].select('a')[0].text
    href = news[i].select('a')[0]['href']
    time = news[i].select('span')[0].text.strip('()')
    
        
    news_dict['title'].append(h2)
    news_dict['href'].append(href)
    news_dict['time'].append(time)
```

```python
news_dict
```

```python
import pandas as pd
news_df = pd.DataFrame(news_dict)[['title','time','href']]
```

```python
from tqdm import tqdm
```

```python
str2 = []
str3 = []
for w in tqdm(range(0,40)):
    try: 
        news_url = news_df.iloc[w,2]
        res = requests.get(news_url)
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text,'html.parser')
        new= soup.select('p')[:-4]
    except:
        print('地址为{}新闻抓取失败'.format(news_url))
        continue
    for i in new:
        str2.append(i.text.strip('\u3000'))
    str3.append(''.join(str2))
    str2= []
```

```python
str3[39]
```

```python
new_dict = {'text':[]}
for t in range(0,len(str3)):
    new_dict['text'].append(str3[t])
```

```python
new_df = pd.DataFrame(new_dict)[['text']]
```
