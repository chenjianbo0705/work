```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import sqlite3
```

```python
news_dict = {'title':[],'href':[],'time':[],'acticle':[]}
for h in range(1,6):
    print(h)
    r = requests.get('http://mil.news.sina.com.cn/roll/index.d.html?cid=57918&page={page}'.format(page = h))
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text,'html.parser')
    new =soup.find_all('li')

    for i in tqdm(new[13:]):
        data = i.select('span')[0].text.strip('()')
        title = i.select('a')[0].text
        href =  i.select('a')[0]['href']
        news_dict['title'].append(title)
        news_dict['href'].append(href)
        news_dict['time'].append(data)
        try:
            res = requests.get(href)
            res.encoding = 'utf-8'
            soup = BeautifulSoup(res.text,'html.parser')
            new = soup.find('div','article')
            t = soup.find('h1','main-title')
            str1=new.text.strip('\n').strip('\u3000')
        except:
            print('地址为{}新闻抓取失败'.format(href))
            
            continue
        news_dict['acticle'].append(str1)
```

```python
df = pd.DataFrame.from_dict(news_dict, orient='index')
df.transpose().head(122)
```

```python
df.to_excel('军事.xlsx')
```

```python
cx = sqlite3.connect(r'D:\Python\获取网络信息用于文本分析\data.db')
pd.io.sql.to_sql(df,'junshi',con = cx)
```
