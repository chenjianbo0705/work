```python
from gensim import corpora, models, similarities
import gensim
#http://radimrehurek.com/gensim/
```

```python
data4 = data1+data2+data3
```

```python
import re
l = []
for i in range(len(data2)):
    reg = "[^\u4e00-\u9fa5]"
    l.append(re.sub(reg,'',data2[i]))
```

```python
import jieba
data5 = []
for  w  in range(len(l)):
    data5.append(jieba.lcut(l[w]))
```

```python
contents_clean = []
all_words = []
for line in data5:
    line_clean = []
    for word in line:
        if word in stopwords:
            continue
        line_clean.append(word)
        all_words.append(str(word))
    contents_clean.append(line_clean)    
```

```python
#做映射，相当于词袋
dictionary = corpora.Dictionary(data5)
corpus = [dictionary.doc2bow(sentence) for sentence in data5]
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=2)#类似Kmeans自己指定K值
```
