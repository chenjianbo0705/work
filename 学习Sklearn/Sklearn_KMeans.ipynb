```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#自己创建数据集
X, y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)   
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as  plt
fig, ax1 = plt.subplots(1)   #建立几个子图
ax1.scatter(X[:, 0], X[:, 1]#.scatter散点图
            ,marker='o' #点的形状
            ,s=8 #点的大小
           )
plt.show()
```

```python
#如果我们想要看见这个点的分布，怎么办？
color = ["red","pink","orange","gray"]
fig, ax1 = plt.subplots(1)

for i in range(4):
    ax1.scatter(X[y==i, 0], X[y==i, 1]
            ,marker='o' #点的形状
            ,s=8 #点的大小
            ,c=color[i]
           )
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = 3
```

```python
cluster = KMeans(n_clusters=n_clusters,random_state=0).fit(X)
```

```python
#重要属性Labels_，查看聚好的类别，每个样本所对应的类
y_pred = cluster.labels_
y_pred
```

```python
#KMeans因为并不需要建立模型或者预测结果，因此我们只需要fit就能够得到聚类结果了
#KMeans也有接口predict和fit_predict，表示学习数据X并对X的类进行预测
#但所得到的结果和我们不调用predict，直接fit之后调用属性labels一模一伴
pre = cluster.fit_predict(X)
pre
```

```python
pre == y_pred#全都是True
```

```python
#重要属性cLuster_centers_，查看质心
centroid = cluster.cluster_centers_
centroid
```

```python
#重要属性inertia_，查看总距离平方和
inertia = cluster.inertia_
inertia
```

```python
color = ["red","pink","orange","gray"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred==i, 0], X[y_pred==i, 1]
            ,marker='o' #点的形状
            ,s=8 #点的大小
            ,c=color[i]
           )
    
ax1.scatter(centroid[:,0],centroid[:,1]   #质心
           ,marker="x"
           ,s=15
           ,c="black")
plt.show()
```

```python
#如果我们把猜测的羡数换成4，Inertia会怎么样？
n_clusters = 4
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
from numpy import * 
import types
def loadDataSet(filename):
    dataMat = []
    fr = open(filename)
    for line in fr.readlines():
        curLine = line.strip().split('\t')
        #map函数是很简单的。意为将第二个参数（一般是数组）中的每一个项，处理为第一个参数的类型
        #fltLine = map(float,curLine)
        fltLine = list(map(float,curLine))
        dataMat.append(fltLine)
    return dataMat
```

```python
dataMat = array(loadDataSet('testSet2.txt'))
```

```python
print(dataMat)
```

```python
import matplotlib.pyplot as  plt
fig, ax1 = plt.subplots(1)   #建立几个子图
ax1.scatter(dataMat[:, 0], dataMat[:, 1]#.scatter散点图
            ,marker='o' #点的形状
            ,s=8 #点的大小
           )
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = 3
```

```python
cluster = KMeans(n_clusters=n_clusters,random_state=0).fit(dataMat)
```

```python
#重要属性Labels_，查看聚好的类别，每个样本所对应的类
y_pred = cluster.labels_
y_pred
```

```python
#重要属性cLuster_centers_，查看质心
centroid = cluster.cluster_centers_
centroid
```
